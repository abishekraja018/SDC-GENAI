```python
# Install necessary libraries
!pip install torch transformers --quiet

# Import necessary libraries
from transformers import BartForConditionalGeneration, BartTokenizer
import torch

# Load the pre-trained BART model and tokenizer for summarization
model_name = "facebook/bart-large-cnn"  # Pre-trained BART model for summarization
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# Set up device for running the model (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to summarize news articles
def summarize_article(article_text):
    # Encode the article text
    inputs = tokenizer.encode("summarize: " + article_text, return_tensors="pt", max_length=1024, truncation=True).to(device)

    # Generate the summary using the BART model
    summary_ids = model.generate(inputs, max_length=150, min_length=50, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode the generated summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Example news article text (replace this with any article)
article_text = """
The latest research shows that AI is revolutionizing many sectors of the economy. From healthcare to finance, artificial intelligence is being applied to optimize processes, reduce costs, and increase efficiency. Companies around the world are investing heavily in AI to stay competitive, and some experts predict that it will be one of the most important technologies of the next decade.
AI is already being used to automate tasks like data entry, customer service, and fraud detection. In healthcare, AI is being applied to assist doctors in diagnosing diseases and suggesting treatments. In finance, AI is being used to predict stock prices and manage risks.
While AI presents significant opportunities, it also poses challenges, particularly around privacy and ethics. Concerns about job displacement and the potential for AI to be used in harmful ways are among the key issues that need to be addressed.
"""

# Call the summarizer function
summary = summarize_article(article_text)

# Display the summary
print("Original Article:")
print(article_text)
print("\nSummary:")
print(summary)
```
