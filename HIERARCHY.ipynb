```python
#  Step 1: Install required libraries
!pip install scipy scikit-learn matplotlib seaborn --quiet

#  Step 2: Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import pandas as pd

#  Step 3: Load dataset
iris = load_iris()
X = iris.data
feature_names = iris.feature_names
df = pd.DataFrame(X, columns=feature_names)

#  Step 4: Perform hierarchical clustering
Z = linkage(X, method='ward')  # You can also try 'complete', 'average', 'single'

#  Step 5: Plot dendrogram (Hierarchy visualization)
plt.figure(figsize=(12, 6))
dendrogram(Z, truncate_mode='lastp', p=30, leaf_rotation=90., leaf_font_size=12., show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index or (Cluster Size)')
plt.ylabel('Distance')
plt.grid(True)
plt.show()

#  Step 6: Form clusters (e.g., 3 clusters)
cluster_labels = fcluster(Z, 3, criterion='maxclust')
df['Cluster'] = cluster_labels

#  Step 7: Visualize clusters using pairplot
sns.pairplot(df, hue='Cluster', palette='viridis')
plt.suptitle("Hierarchical Clustering (Ward's Method)", y=1.02)
plt.show()

#  Step 8: Preview the data
print(df.head())
```
