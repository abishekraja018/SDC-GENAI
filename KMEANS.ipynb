```python
# 📦 Step 1: Install required libraries
!pip install matplotlib scikit-learn --quiet

# 📚 Step 2: Import libraries
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import pandas as pd

# 🌸 Step 3: Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# 📊 Step 4: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# 🧠 Step 5: Add the predicted cluster to the data
df = pd.DataFrame(X, columns=feature_names)
df['Predicted Cluster'] = labels
df['True Label'] = y

# 🎨 Step 6: Reduce dimensions for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# 🎯 Step 7: Plot the clusters
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=60)
plt.title("K-Means Clustering on Iris Dataset (2D PCA Projection)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.colorbar(scatter, label='Cluster')
plt.grid(True)
plt.show()

# 📝 Step 8: Show first few results
print(df.head())
```
